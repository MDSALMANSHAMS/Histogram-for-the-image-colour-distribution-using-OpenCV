# Histogram for the image colour distribution

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread("F:\Image\Cat\cat1.jpg")
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('Gray', gray)

blank = np.zeros(img.shape[:2], dtype='uint8')
mask = cv.circle(blank, (img.shape[1]//2,img.shape[0]//2), 50, 255, -1)
masked = cv.bitwise_and(gray,gray,mask=mask)
cv.imshow('Mask', masked)

cv.waitKey(0)
```

### 1. Grayscale histogram

```python
gray_hist = cv.calcHist([gray], [0], masked, [256], [0,256] )

plt.figure()
plt.title('Grayscale Histogram')
plt.xlabel('Bins')
plt.ylabel('# of pixels')
plt.plot(gray_hist)
plt.xlim([0,256])
plt.show()
```

### 2. Colour Histogram

```python
masked_1 = cv.bitwise_and(img,img,mask=mask)
cv.imshow('Mask', masked_1)
colors = ('b', 'g', 'r')

plt.figure()
plt.title('Colour Histogram')
plt.xlabel('Bins')
plt.ylabel('# of pixels')

for i,col in enumerate(colors):
    hist = cv.calcHist([img], [i], mask, [256], [0,256])
    plt.plot(hist, color=col)
    plt.xlim([0,256])
    
plt.show()
cv.waitKey(0)
```
